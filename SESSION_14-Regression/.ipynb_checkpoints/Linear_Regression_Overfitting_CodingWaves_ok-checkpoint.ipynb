# Linear regression using the Normal Equation

```python
X_sq = np.random.uniform(2, 500, size=(100, 1))
y = 200 + 3 * X_sq + np.random.randn(100, 1)*50
X_sq_list = X_sq.ravel().tolist()
```

```python
X_b = np.c_[np.ones((100, 1)), X_sq]  # add x0 = 1 to each instance
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

# Linear regression using batch gradient descent

```python
def  cal_cost(theta,X,y):
    '''
    
    Calculates the cost for given X and Y. The following shows and example of a single dimensional X
    theta = Vector of thetas 
    X     = Row of X's np.zeros((2,j))
    y     = Actual y's np.zeros((2,1))
    
    where:
        j is the no of features
    '''
    
    m = len(y)
    
    predictions = X.dot(theta)
    cost = (1/2*m) * np.sum(np.square(predictions-y))
    return cost

def gradient_descent(X,y,theta,learning_rate=0.01,iterations=100):
    '''
    X    = Matrix of X with added bias units
    y    = Vector of Y
    theta=Vector of thetas np.random.randn(j,1)
    learning_rate 
    iterations = no of iterations
    
    Returns the final theta vector and array of cost history over no of iterations
    '''
    m = len(y)
    cost_history = np.zeros(iterations)
    theta_history = np.zeros((iterations,2))
    for it in range(iterations):
        
        prediction = np.dot(X,theta)
        errors_vector = prediction - y
        
        theta = theta - ((2/m)*(X.T.dot(errors_vector)))* learning_rate
        theta_history[it,:] = theta.T
        cost_history[it]  = cal_cost(theta,X,y)
        
    return theta, cost_history, theta_history
```

```python
lr =0.001
n_iter = 50000

np.random.seed(0)

X = np.random.uniform(2, 50, size=(1000, 1))
y = 200 + 3 * X + np.random.randn(1000,1) * 20

theta = np.random.randn(2,1)

X_b = np.c_[np.ones((len(X),1)),X]

theta,cost_history,theta_history = gradient_descent(X_b,y,theta,lr,n_iter)
```

Your task is fitting the data with a Polynomial Regression, building 5 models with different polynomial expansion.<br>
The degrees to use are in array "degrees". Use a for loop to fit the data with different degrees. Use PolynomialFeatures(degree=..., include_bias=False)

```python
from sklearn.preprocessing import PolynomialFeatures
degrees = np.arange(2,30,5)
```

```python
plt.figure(figsize=(16,9))
for deg in degrees:
    # Calculate ploy_features
    poly_features = PolynomialFeatures(degree=deg, include_bias=False)
    X_poly = poly_features.fit_transform(X)
    # Fit the data with Linear Regression
    lin_reg = LinearRegression()
    lin_reg.fit(X_poly, y)
    
    # Plot function: DONT TOUCH IT PLEASE
    X_new=np.linspace(-3, 3, 100).reshape(100, 1)
    X_new_poly = poly_features.transform(X_new)
    y_new = lin_reg.predict(X_new_poly)
    plt.plot(X, y, "b.")
    plt.plot(X_new, y_new, linewidth=2, label=deg)
    plt.xlabel("$x_1$", fontsize=18)
    plt.ylabel("$y$", rotation=0, fontsize=18)
    plt.axis([-3, 3, 0, 10])
    plt.legend()
plt.grid()
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge

np.random.seed(42)
m = 20
X = 3 * np.random.rand(m, 1)
y = 1 + 0.5 * X + np.random.randn(m, 1) / 1.5
X_new = np.linspace(0, 3, 100).reshape(100, 1)

def plot_model(model_class, degree, alphas, **model_kargs):
    for alpha, style in zip(alphas, ("b-", "g--", "r:")):
        model = model_class(alpha, **model_kargs) if alpha > 0 else LinearRegression()
        
        if degree is not None:
            model = Pipeline([
                    ("poly_features", PolynomialFeatures(degree=10, include_bias=False)),
                    ("std_scaler", StandardScaler()),
                    ("regul_reg", model),
                ])
        model.fit(X, y)
        y_new_regul = model.predict(X_new)
        
        lw = 2 if alpha > 0 else 1
        plt.plot(X_new, y_new_regul, style, linewidth=lw, label=r"$\alpha = {}$".format(alpha))
    plt.plot(X, y, "b.", linewidth=3)
    plt.legend(loc="upper left", fontsize=15)
    plt.xlabel("$x_1$", fontsize=18)
    plt.axis([0, 3, 0, 4])
    return model

plt.figure(figsize=(8,4))

plt.subplot(121)
plot_model(Ridge, degree=None, alphas=(0, 10, 100), random_state=42)
plt.ylabel("$y$", rotation=0, fontsize=18)

plt.subplot(122)
plot_model(Ridge, degree=30, alphas=(0, 10**-5, 1), random_state=42)

plt.show()
```

## Lasso Regularization

```python
from sklearn.linear_model import Lasso

plt.figure(figsize=(8,4))
plt.subplot(121)
plot_model(Lasso, degree=None, alphas=(0, 0.1, 1), random_state=42)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.subplot(122)
plot_model(Lasso, degree=10, alphas=(0, 10**-7, 1), tol=1, random_state=42)

# save_fig("lasso_regression_plot")
plt.show()
```
