# Classification - Click on internet Advertisement

In this project we will be working with advertising data set, `indicating whether or not a particular internet user clicked on an Advertisement`. We will try to create a model that will predict whether or not they will click on an ad based off the features of that user.

This data set contains the following features:

* 'Daily Time Spent on Site': consumer time on site in minutes
* 'Age': cutomer age in years
* 'Area Income': Avg. Income of geographical area of consumer
* 'Daily Internet Usage': Avg. minutes a day consumer is on the internet
* 'Ad Topic Line': Headline of the advertisement
* 'City': City of consumer
* 'Male': Whether or not consumer was male
* 'Country': Country of consumer
* 'Timestamp': Time at which consumer clicked on Ad or closed window
* 'Clicked on Ad': 0 or 1 indicated clicking on Ad

# Train and Test Split

Now it's time to do a train test split, and train our model!

You'll have the freedom here to choose columns that you want to train on!

** Split the data into training set and testing set using train_test_split**

```python
from sklearn.model_selection import train_test_split
```

```python
X = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]
y = ad_data['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Please print the shapes of X_train and X_test
print(X_train.shape)
print(X_test.shape)
```

** Train and fit a logistic regression model on the training set.**

## Try to use Logistic Regression Classification Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression(solver='lbfgs')
logmodel.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_validate
scores_log_reg = cross_validate(LogisticRegression(solver='lbfgs'), 
                                X=X_train, y=y_train, cv=5, scoring='accuracy')
```

```python
scores_log_reg = pd.DataFrame(scores_log_reg)
scores_log_reg
```

## Predictions on test set and Evaluations of the model
** Now predict values for the test set.**

```python
predictions = logmodel.predict(X_test)
```

** Create the classification report**

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

## Try with: Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
randomForest = RandomForestClassifier(n_estimators=300, max_depth=10)
randomForest.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_validate
scores_rand_forest = cross_validate(randomForest, X=X_train, y=y_train, cv=5, scoring='accuracy')
```

```python
scores_rand_forest = pd.DataFrame(scores_rand_forest)
```

```python
scores_rand_forest.mean(axis=0)
```

# Random Forest Prediction on test set

```python
print(classification_report(y_test, randomForest.predict(X_test)))
```

# Ok! Seems that RandomForest works best! Let's Fine Tune it

### Fine tune your Random Forest Classifier

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(RandomForestClassifier(), param_grid={'max_depth': range(5, 100, 10), 
                                                         'n_estimators': range(1,100,10)})
```

```python
grid.fit(X_train, y_train)
```

```python
pd.DataFrame(grid.cv_results_)[['param_max_depth', 'params', 'rank_test_score', 'mean_test_score']].\
sort_values(by='rank_test_score')
```

## Predictions and Evaluations
** Now predict values for the testing data.**

```python
predictions = grid.predict(X_test)
```

** Create a classification report for the model.**

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

## Features Importance with Random Forest

# Bonus Track!

```python
from sklearn.preprocessing import StandardScaler
def data_preparation(df):
    # the goal of this function is to return a StandardScaler Object fitted with X_train data
    # select only numerical columns, except the target
    df = df.select_dtypes(np.number)
    st = StandardScaler()
    st.fit(df)
    return st
```

```python
logmodel = LogisticRegression(solver='lbfgs', C=1)
```

```python
scaled_train = data_preparation(X_train).transform(X_train)
scores_logmodel = cross_validate(logmodel, 
                                 X=scaled_train, 
                                 y=y_train, 
                                 cv=5, 
                                 scoring='accuracy')
```

```python
pd.DataFrame(scores_logmodel).mean()
```

```python
# NON scaled data
logmodel.fit(X_train, y_train)
preds = logmodel.predict(X_test)
print(classification_report(y_test, preds))
```

```python
# Scaled data
scaled_X_train = data_preparation(X_train).transform(X_train)
scaled_X_test = data_preparation(X_train).transform(X_test)

logmodel.fit(scaled_X_train, y_train)
preds = logmodel.predict(scaled_X_test)
print(classification_report(y_test, preds))
```

```python
# Roc curve
from sklearn.metrics import roc_curve
plt.plot(roc_curve(y_test, preds, pos_label=1)[1])
```

```python
preds_probas = logmodel.predict_proba(scaled_X_test)
```

```python
temp = pd.DataFrame(data={'probas': preds_probas[:, 1], 'real_class': y_test})

sns.distplot(temp[temp['real_class']==0]['probas'], kde=True, hist=False)
sns.distplot(temp[temp['real_class']==1]['probas'], kde=True, hist=False)
```
