# Data Analysis and Visualization

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
%matplotlib inline
```

### Read Dataset

```python
df = pd.read_csv('../Data/Movie-Ratings.csv')
df.head()
```

### Dataset info

```python
df.info()
```

### Rename the columns

```python
# Get column names
df.columns
```

```python
# rename columns
df.columns = ['Film', 'Genre', 'Critic_Ratings', 'Audience_Ratings','Budget_Ml$', 'Year']
```

```python
# check the changes
df.columns
```

### Plotting the Audiance ratings against the Critic Ratings
Scatter plot

```python
# set fig size
plt.figure(figsize=(14, 6))

# plot function
sns.scatterplot(data= df, x='Critic_Ratings', y='Audience_Ratings',)

# plot title
plt.title("Audience Rating vs Critic Ratings")

# show
plt.show()
```

#### Compute the Correlation

```python
# check the correlation between the numeric variable
df[['Critic_Ratings','Audience_Ratings','Budget_Ml$']].corrwith(df.Critic_Ratings)
```

### Plotting the Audiance ratings against the Critic Ratings colored by Genre
Scatter plot

```python
# fig size
fig = plt.figure(figsize=(14,6))

# plot function
sns.scatterplot(data=df, x='Critic_Ratings', y='Audience_Ratings', hue='Genre')

# show
plt.show()
```

```python
df.head()
```

### Plotting the Audiance Ratings against the Critic Ratings colored by Genre and mapped by Budget_Ml$
Scatter plot

```python
# fig size
fig = plt.figure(figsize=(14,6))

# plot function
sns.scatterplot(data=df, x='Critic_Ratings', y='Audience_Ratings', hue='Genre', size='Budget_Ml$')

# show
plt.show()
```

### Plotting the Audience Rating by Genre
Categorical plot

```python
# plot function (kind=box)
sns.catplot(data=df, x='Genre', y='Audience_Ratings', kind='box', height=5, aspect=1.7)

# plot title
plt.title("Audience Rating by Genre")

# show
plt.show()
```

### Plotting Budget Distribution

Univariate distribution plot

```python
# set the fig size
plt.figure(figsize=(10,4))
sns.distplot(df['Budget_Ml$'])
plt.show()
```

```python
# set the facet grid
g = sns.FacetGrid(data=df, hue='Genre', height=4, aspect=3) #HINT: Try row='Genre'

# map the faced grid with function
g.map(sns.kdeplot, 'Budget_Ml$', shade=True)

# Adding the legend
g.add_legend()

# plot title
plt.show()
```

### Budget spent per every movie evary year

Categorical plot

```python
# plot finction
sns.catplot(data=df, x='Year', y='Budget_Ml$', ci=None, height=5, aspect=1.7)

# show
plt.show()
```

### Budget spent per every movie evary year

Categorical plot **(hue='Genre')**

```python
# plot finction
sns.catplot(data=df, x='Year', y='Budget_Ml$', hue='Genre', ci=None, height=5, aspect=1.7)

# show
plt.show()
```

### Average of the badget spent for every year

Categorical plot **(kind='point')**

```python
sns.catplot(data=df, x='Year', y='Budget_Ml$', kind='point', ci=None, height=4, aspect=2, color='green')
plt.show()
```

#### Another way to visualize the budget average per year
The **bar plot:** Represents an **estimate of central tendency for a numeric variable** with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars.

```python
# fig setting
plt.figure(figsize=(10,5))

# plot function
sns.barplot(data=df, x='Year', y='Budget_Ml$') # estimator = mean

# plot title
plt.title('Budget mean per year per movie')

# show
plt.show()
```

### Total budget spent every year

Bar plot

```python
# fig setting
plt.figure(figsize=(10,5))

# plot function (sum estimator)
sns.barplot(data=df, x='Year', y='Budget_Ml$', estimator=np.sum)

# plot title
plt.title('Budget per Year')

# show
plt.show()
```

### The number of movies  made every hear

```python
# plot function (kind)
sns.catplot(data=df, x='Year', kind='count', palette='viridis', height=5, aspect=2)

# show
plt.show()
```

### Total budget sent every year for every genre

Bar plot

```python
plt.figure(figsize=(15,5))
sns.barplot(data=df, x='Year', y='Budget_Ml$',hue='Genre',ci=None, estimator=np.sum)
plt.title('Budget per Year')
plt.show()
```

```python
sns.lmplot(data=df, x='Critic_Ratings', y='Audience_Ratings', ci=None)
```

```python
#fig = plt.figure(figsize=(14,6))
sns.lmplot(data=df, x='Critic_Ratings', y='Audience_Ratings', ci=None, col='Genre', col_wrap=2)
plt.show()
```
