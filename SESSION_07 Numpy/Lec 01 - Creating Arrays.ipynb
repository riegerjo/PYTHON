## The .arange() method

#### 1-D array

```python
# Using the .arange() method you will get a 1-D array by default

array_1d = np.arange(10)
array_1d
```

```python
# start, stop and step

np.arange(5, 50, 5)
```

#### 2-D array, the reshape method

```python
# Using the reshape method

array_2d = np.arange(5, 50, 5).reshape(3,3)
array_2d
```

```python
# Another example...
np.arange(16).reshape(4,4)
```

#### 3-D array

```python
# reshape(0=Z, 1=Y, 2=X)
array_3d = np.arange(24).reshape(4,3,2)
array_3d
```

### Linespace method

When arange is used with floating point arguments, it is generally not possible to predict the number of elements obtained, due to the finite floating point precision. 

For this reason, it is usually better to use the function **linspace** that receives as an argument the number of elements that we want, instead of the step:

```python
np.arange( 10, 30, 5 )
```

```python
np.arange( 0, 3, 0.3 )                 # it accepts float arguments
```

We can chose how many numbers we want to set in a given range

```python
# linespace method()
np.linspace(0, 3, 8)
```

## The randoms methods

#### random.rand()
Create an array of the given shape and populate it with random samples from a uniform distribution over [0, 1).

In probability theory and statistics, the [continuous uniform distribution](https://en.wikipedia.org/wiki/Uniform_distribution_(continuous)) or rectangular distribution is a family of symmetric probability distributions such that for each member of the family, all intervals of the same length on the distribution's support are equally probable.

```python
np.random.rand(5)
```

#### random.randn()

Return a sample (or samples) from the [standard normal distribution](https://en.wikipedia.org/wiki/Normal_distribution).

```python
np.random.randn(10)
```
