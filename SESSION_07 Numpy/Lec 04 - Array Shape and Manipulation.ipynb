#### Create array

```python
#Create array
arr = np.arange(50).reshape((10,5))

#Show
arr
```

### Reshape vs Resize 

 * The reshape method returns an array containing the same data with a new shape **not in-place**.

 * The resize method returns an array containing the same data with a new shape **in-place**.

```python
# Reshape (not in-place function)
arr.reshape((5,10))
```

```python
# If I check the original array it didn't change the shape
arr
```

```python
# Resize (in-place method)
arr.resize((5,10))
```

```python
# After resize, if we check the original array it changed the shape
arr
```

#### Flatten array

Return a copy of the array collapsed into **one dimension**.

```python
# Convert 2-D array into 1-D array
ft = arr.flatten()
```

```python
ft
```

```python
# Check the shape
ft.shape
```

### We can transpose a 3d array

```python
from IPython.display import Image
Image(filename='../Images/np3darray.PNG') 
```

```python
# For 3D matrix
arr3d = np.arange(24).reshape((4,3,2)) # Axis Index:(0, 1, 2) Dimensions: 0=z, 1=y, 2=x

#Show
arr3d
```

```python
# just to understand the axis
arr3d.sum(axis=2)
```

```python
arr3d.shape
```

This means that: 
* On the axis 0=Z we have 4 sub-elements;
* On the axis 1=Y we have 3 sub-elements;
* On the axis 2=X we have 2 sub-elements;

```python
# If we don't specify any parameter transposition will be... (2,3,4)

arr3d.transpose()  
```

```python
# Axis 1 on Z dimension
# Axis 0 on Y dimension
# Axis 2 on X dimension


arr3d.transpose((1,0,2))
```
